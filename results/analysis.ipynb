```python
import json
import pandas as pd
pd.get_option("display.max_columns")


with open('../results/hadoop_results_final.json') as d:
    data = json.load(d)
    
```

```python
df = pd.json_normalize(data)
```

```python
df = df.rename(columns = { i: "PC-" + i.split('.')[1] for i in  df.columns if i.startswith("Parameter Calibration")})
df = df.rename(columns = { i: "BC-" + i.split('.')[1] for i in  df.columns if i.startswith("Bloom Filter Creation")})
df = df.rename(columns = { i: "PV-" + i.split('.')[1] for i in  df.columns if i.startswith("Parameter Validation")})
```

```python
final_table_columns = ['falsePositiveRate', 'nReducerJob0', 'nReducerJob1', 'nReducerJob2', 'nLineSplitJob1', 'executionTime',
                      'PV-CPU time spent (ms)', 'BC-CPU time spent (ms)', 'PC-CPU time spent (ms)']
df = df.drop(columns=[col for col in df if col not in final_table_columns])
df
```

```python

df = df.groupby(['falsePositiveRate', 'nReducerJob0', 'nReducerJob1', 'nReducerJob2', 'nLineSplitJob1']).agg(['mean', 'std'])
df = df.reset_index(['falsePositiveRate', 'nReducerJob0', 'nReducerJob1', 'nReducerJob2', 'nLineSplitJob1'])
df.columns = ['_'.join(col) for col in df.columns.values]
df
```

```python
df.sort_values(by=[('BC-CPU time spent (ms)_mean')])
df.sort_values(by=[('BC-CPU time spent (ms)_mean')]).drop_duplicates(['nReducerJob1_', 'nLineSplitJob1_'])
```
